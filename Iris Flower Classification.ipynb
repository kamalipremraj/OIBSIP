### Preprocessing

#### Feature Scaling
To keep the values between -1 to 1


```python
from sklearn.preprocessing import StandardScaler
```

```python
model=StandardScaler()
```

```python
X=df.drop(["Species"],axis=1)
y=df["Species"]
```

```python
scaler=model.fit(X)
```

```python
X_trans=scaler.transform(X)
```

```python
X_trans
```

#### Label Encoding
Change the string to numerical value ( caregorical value)

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# Changing the y value "Species" to numeric
encoder=LabelEncoder()

y=encoder.fit_transform(y)
```

```python
y
```

#### Splitting

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
#X_train.value_counts
#X_test.value_counts
## To count/get the values
```

### Model Selection

#### SupportVectorClassifier

```python
from sklearn.svm import SVC
```

```python
# In svc we take linear model (our data is linearly seperatable) for non-linear data we go for gradient
svc=SVC(kernel='linear')
```

```python
svc.fit(X_train,y_train)
```

```python
y_pred=svc.predict(X_test)
```

```python
y_pred
```

#### Metrics
To find model accuracy

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_pred,y_test)
```

#### LogisticRegression

```python

from sklearn.linear_model import LogisticRegression
lg=LogisticRegression()
```

```python
lg.fit(X_train,y_train)
```

```python
y_predict=lg.predict(X_test)
accuracy_score(y_predict,y_test)
```

#### KNeighborsClassifier

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train,y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
accuracy_score(y_pred,y_test)
```
